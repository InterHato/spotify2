```python
# Shows a user's playlists (need to be authenticated via oauth)
from collections import Counter
import sys
import spotipy
import spotipy.util as util
import matplotlib.pyplot as plt
import random
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from scipy.stats import gaussian_kde


username = ''

client_id = ""
client_secret = ""
redirect_uri = ""




scope = 'user-library-read user-top-read playlist-modify-public user-follow-read'
token = util.prompt_for_user_token(username, scope, client_id, client_secret, redirect_uri)



    
    #Step 1. Authenticating Spotipy

def authenticate_spotify():
    print('...connecting to Spotify')
    sp = spotipy.Spotify(auth=token)
    print("we")
    return sp




def user_playlist_tracks_full(spotify_connection, user):
    """ Get full details of the tracks of a playlist owned by a user.
        https://developer.spotify.com/documentation/web-api/reference/playlists/get-playlists-tracks/

        Parameters:
            - user - the id of the user
            - playlist_id - the id of the playlist
            - fields - which fields to return
            - market - an ISO 3166-1 alpha-2 country code.
    """

    # first run through also retrieves total no of songs in library
    response = spotify_connection.current_user_saved_tracks(limit=20,offset=0)
    print(len(response))
    #print(type(response))
    results = response["items"]
    print(len(results))
    #print("1")
    #print(response["total"])
    # subsequently runs until it hits the user-defined limit or has read all songs in the library
    #response["total"]
    counter=0
    while len(results) < response["total"]:
    
        #print("here")
        response = spotify_connection.current_user_saved_tracks(limit=20,offset=len(results))
        results.extend(response["items"])
        counter+=1
        print(counter)
        #print(results)
    print(len(results))
    uri_list=[]
    dates=[]
    audio_features=[]
    for song in results:
        #print(song)
        #print(len(song))
        uri_list.append((song['track'])['uri'])
        dates.append(song['added_at'])
        #print(song['added_at'])
    #print(len(uri_list)) 
    #print(len(dates))
        
    #print(results[0])
    
    #mine= (results[0])['track']['uri']

    #print (mine)
    cleandates=[]
    for date in dates:
    
        cleandates.append(date[0:10])

    #print(cleandates)
    #print(type(cleandates))
    x = [dt.datetime.strptime(d, '%Y-%m-%d').date() for d in cleandates]
        
   # print(x)
       
    for uri in uri_list:
            
        #print(uri)
        [audio_feature_no_bracket]= spotify_connection.audio_features(uri)
        audio_features.append(audio_feature_no_bracket)
            
        
    print(len(audio_features))
        
        
    y=[]
        
    for song in audio_features:
            
        y.append(song['valence'])
    #print(y)
    #print(len(y))
    #plt.scatter(y,y)
   
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator())
    print(len(x))
    print(len(y))
    
   
    
    
    
    
    
    plt.scatter(x,y)
    print(x)
    print(y)
    plt.gcf().autofmt_xdate()  
    print(x[0])
    print(x[-1])

    #print(len(Counter(x).keys()))
    #print(len(Counter(y).keys()))
    print(len(list(Counter(x).keys())))
    print(len(list(Counter(y).keys())))
    # do a density  shctic colour thing to compare that in teh top right corner less songs





user_playlist_tracks_full(spotify_connection= authenticate_spotify(), user='')
```
