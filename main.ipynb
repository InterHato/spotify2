```python
# Shows a user's playlists (need to be authenticated via oauth)
from collections import Counter
import sys
import spotipy
import spotipy.util as util
import matplotlib.pyplot as plt
import random
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
from scipy.stats import gaussian_kde
import pandas as pd




username = ''
client_id = ""
client_secret = ""
redirect_uri = ""

user=''


scope = 'user-library-read user-top-read playlist-modify-public user-follow-read'
token = util.prompt_for_user_token(username, scope, client_id, client_secret, redirect_uri)



    
    #Step 1. Authenticating Spotipy

def authenticate_spotify():
    print('...connecting to Spotify')
    sp = spotipy.Spotify(auth=token)
    print("we")
    return sp

spotify_connection = authenticate_spotify()



    
response = spotify_connection.current_user_saved_tracks(limit=20,offset=0)
print(len(response))
#print(type(response))
results = response["items"]
print(len(results))
    
#response["total"]
counter=0
while len(results) < response["total"]:
    
        
    response = spotify_connection.current_user_saved_tracks(limit=20,offset=len(results))
    results.extend(response["items"])
    counter+=1
    print(counter)
    #print(results)
print(len(results))
uri_list=[]
dates=[]
audio_features=[]
for song in results:
    #print(song)
    #print(len(song))
    uri_list.append((song['track'])['uri'])
    dates.append(song['added_at'])
        
#print(len(uri_list)) 
#print(len(dates))
        
#print 
cleandates=[]
for date in dates:
    
    cleandates.append(date[0:10])

#print(cleandates)
#print(type(cleandates))
x = [dt.datetime.strptime(d, '%Y-%m-%d').date() for d in cleandates]
        
# print(x)
       
for uri in uri_list:
            
    #print(uri)
    [audio_feature_no_bracket]= spotify_connection.audio_features(uri)
    audio_features.append(audio_feature_no_bracket)
            
        
print(len(audio_features))
        






```
